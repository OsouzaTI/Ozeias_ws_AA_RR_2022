```python
# importando as bibliotecas
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# carregando arquivo de tempo de execução
tempos = pd.read_csv('exec_time.csv', sep=';')

ordenado = pd.read_csv('execution_time_ordered_numbers.csv', sep=';')
aleatorio = pd.read_csv('execution_time_sorted_numbers.csv', sep=';')
inversamenteOrdenado = pd.read_csv('execution_time_inverse_ordered_numbers.csv', sep=';')

ordenadoMedio = pd.read_csv('execution_time_ordered_numbers_medio.csv', sep=';')
aleatorioMedio = pd.read_csv('execution_time_sorted_numbers_medio.csv', sep=';')
inversamenteOrdenadoMedio = pd.read_csv('execution_time_inverse_ordered_numbers_medio.csv', sep=';')
```

```python
# tempos.head()
# ordenado.head()
```

```python
# valores de x
tx = np.linspace(0, 10)
def funcaoCusto(x):
    return (1/2) * (x**2 - x)

def complexidade(x):
    return x**2

plt.figure(figsize=(8, 6))
plt.title('Comportamento da função O(n) = n^2\n 0 ≤ n ≤ 10')
plt.plot(tx, complexidade(tx), color='blue', label='O(n) = n^2')
plt.grid(True)
plt.legend()
plt.savefig('complexidade_n2.png')
```

```python
plt.figure(figsize=(8, 6))
plt.title('Análise comportamental')
plt.plot(tempos['n'], tempos['time'])
plt.xlabel('Tamanho da entrada')
plt.ylabel('Tempo em segundos')
plt.grid(True)
plt.savefig('analise_algoritmo.png')
plt.show()
```

```python
# Segunda execução de testes
plt.figure(figsize=(8, 6))
plt.title('Gráfico do tempo em função da entrada')
plt.plot(ordenado['n'], ordenado['time'], label='Numeros Ordenados')
plt.plot(aleatorio['n'], aleatorio['time'], label='Numeros Aleatórios')
plt.plot(inversamenteOrdenado['n'], inversamenteOrdenado['time'], label='Inversamente ordenado')
plt.xlabel('Tamanho da entrada')
plt.ylabel('Tempo em segundos')
plt.grid(True)
plt.legend()
plt.savefig('analise_algoritmo_ordenado.png')
plt.show()
```

```python
# terceira execução de testes
plt.figure(figsize=(8, 6))
plt.title('Gráfico do tempo médio de execução')
plt.plot(ordenadoMedio['n'], ordenadoMedio['time'], label='Numeros Ordenados')
plt.plot(aleatorioMedio['n'], aleatorioMedio['time'], label='Numeros Aleatórios')
plt.plot(inversamenteOrdenadoMedio['n'], inversamenteOrdenadoMedio['time'], label='Inversamente ordenado')
plt.xlabel('Tamanho da entrada')
plt.ylabel('Tempo em segundos')
plt.grid(True)
plt.legend()
plt.savefig('analise_algoritmo_medio.png')
plt.show()
```

```python
sample = tempos.sample(frac=0.4).sort_index()

plt.figure(figsize=(6, 4))
# plt.title('Gráfico de tendência')
plt.scatter(sample['n'], sample['time'], s=10, marker='.')
plt.xlabel('Tamanho da entrada')
plt.ylabel('Tempo em segundos')
plt.grid(True)

# Calculo da linha de tendência
z = np.polyfit(sample['n'], sample['time'], 2)
p = np.poly1d(z)
plt.plot(sample['n'], p(sample['n']), "r--", label='Linha de tendência')
plt.savefig('linha_tendencia.jpg')
plt.legend()
```
